```python
import numpy as np
import pandas as pd
np.int = int
```

```python
!git clone https://github.com/amir-saniyan/HodaDatasetReader.git

import matplotlib.pyplot as plt
from HodaDatasetReader.HodaDatasetReader import read_hoda_dataset
```

```python
X_train, y_train = read_hoda_dataset('/content/HodaDatasetReader/DigitDB/Train 60000.cdb')
X_test, y_test   = read_hoda_dataset('/content/HodaDatasetReader/DigitDB/Test 20000.cdb')
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python


plt.figure(figsize=(12,5))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i].reshape(32,32))
    plt.title(y_train[i])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat  = X_test.reshape(X_test.shape[0], -1)

print("Train shape:", X_train_flat.shape)
print("Test shape :", X_test_flat.shape)
```

```python
svm_model = SVC(kernel='rbf', gamma=0.01, C=10)
```

```python
svm_model.fit(X_train_flat, y_train)
```

```python
y_pred = svm_model.predict(X_test_flat)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix - Hoda Dataset (SVM)")
plt.show()
```

```python
svm_model_l_100 = SVC(kernel='linear', gamma=0.01, C=100)
```

```python
svm_model_l_100.fit(X_train_flat, y_train)
```

```python
y_pred = svm_model_l_100.predict(X_test_flat)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
